```python
import pandas as pd
import matplotlib.pyplot as plt
import os

#os.chdir('../')
print("Current working directory:", os.getcwd())

# Load the CSV file
file_path = 'Prosperity3Enigma/data_25/Round1/prices_round_1_day_-1.csv'  # Replace with your actual file path or upload mechanism
print(os.path.exists(file_path))
df = pd.read_csv(file_path, sep=';')

# Ensure timestamp is sorted properly
df = df.sort_values(by=['product', 'timestamp'])

# Get unique products
products = df['product'].unique()

# Plot mid_price for each product
for product in products:
    subset = df[df['product'] == product]
    plt.figure()
    plt.plot(subset['timestamp'], subset['mid_price'], marker='o')
    plt.title(f'Mid Price over Time for {product}')
    plt.xlabel('Timestamp')
    plt.ylabel('Mid Price')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot profit_and_loss for each product
for product in products:
    subset = df[df['product'] == product]
    plt.figure()
    plt.plot(subset['timestamp'], subset['profit_and_loss'], marker='o')
    plt.title(f'Profit and Loss over Time for {product}')
    plt.xlabel('Timestamp')
    plt.ylabel('Profit and Loss')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
